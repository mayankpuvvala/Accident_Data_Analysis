```python
# List of HTML files to merge
html_files = ['Accused_Details.html', 'AccFir.html', 'Accident_Reports.html', 'Accident_Complainant.html']

# Open the merged HTML file
with open('Main3.html', 'w', encoding='utf-8') as merged_file:
    # Write the opening HTML tags
    merged_file.write('<html>\n<head>\n<style>\n')
    
    # Add CSS for styling
    merged_file.write('body { font-family: Arial, sans-serif; }\n')
    merged_file.write('.section { margin-bottom: 20px; }\n')
    merged_file.write('.section-heading { margin-left: 20px; border-left: 2px solid #ccc; padding-left: 10px; }\n')
    merged_file.write('.toc { position: fixed; top: 50px; right: 10px; background-color: #f0f0f0; border: 1px solid #ccc; padding: 10px; display: none; }\n')
    merged_file.write('.toc ul { list-style-type: none; padding-left: 0; }\n')
    merged_file.write('.toc li { margin-bottom: 5px; }\n')
    merged_file.write('.show-toc { position: fixed; top: 10px; right: 10px; background-color: #f0f0f0; border: 1px solid #ccc; padding: 5px; cursor: pointer; }\n')
    merged_file.write('</style>\n')
    merged_file.write('<script>\n')
    merged_file.write('function toggleTOC() {\n')
    merged_file.write('var toc = document.querySelector(".toc");\n')
    merged_file.write('if (toc.style.display === "none") {\n')
    merged_file.write('toc.style.display = "block";\n')
    merged_file.write('} else {\n')
    merged_file.write('toc.style.display = "none";\n')
    merged_file.write('}\n')
    merged_file.write('}\n')
    merged_file.write('</script>\n')
    merged_file.write('</head>\n<body>\n')
    
    # Write the button to show/hide the TOC
    merged_file.write('<div class="show-toc" onclick="toggleTOC()">Show TOC</div>\n')
    
    # Write the table of contents
    merged_file.write('<div class="toc">\n')
    merged_file.write('<ul>\n')
    for html_file in html_files:
        file_name = html_file.split('.')[0]  # Extract file name without extension
        # Get the title of the HTML file (assuming it's the first h1 tag)
        with open(html_file, 'r', encoding='utf-8') as current_file:
            html_content = current_file.read()
            title_index = html_content.find('<h1>') + len('<h1>')
            title_end_index = html_content.find('</h1>')
            title = html_content[title_index:title_end_index]
        merged_file.write(f'<li><a href="#{file_name}">{title}</a></li>\n')
    merged_file.write('</ul>\n')
    merged_file.write('</div>\n')
    
    # Loop through each HTML file
    for html_file in html_files:
        # Open the current HTML file
        with open(html_file, 'r', encoding='utf-8') as current_file:
            # Read all lines of the HTML file
            html_content = current_file.read()
            
            # Write the section with file title as the ID
            file_name = html_file.split('.')[0]  # Extract file name without extension
            # Get the title of the HTML file (assuming it's the first h1 tag)
            title_index = html_content.find('<h1>') + len('<h1>')
            title_end_index = html_content.find('</h1>')
            title = html_content[title_index:title_end_index]
            
            # Find the index of the Overview section
            overview_index = html_content.find('<h2>Overview</h2>')
            
            # Write the section before the Overview
            merged_file.write(f'<div id="{file_name}" class="section">\n')
            merged_file.write(f'<div class="section-heading">{title}</div>\n')
            merged_file.write(html_content[:overview_index])
            merged_file.write('</div>\n')
            
            # Write the Overview section
            merged_file.write(html_content[overview_index:])
    
    # Write the closing HTML tags
    merged_file.write('</body>\n</html>')
```
